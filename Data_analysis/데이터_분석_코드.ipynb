```python
import pandas as pd
import json 
import math
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='white') 
rcParams["figure.figsize"] = 12, 6
```

```python
data = pd.read_csv("C:/Users/gimm7/OneDrive/바탕 화면/위험도_데이터_10m.csv", encoding='cp949')
data.drop(["출처"], axis=1, inplace=True)
data.head()
```

```python
data.drop(['노드1', '노드2', '거리','공원', '위험도', '골목상권','공공체육시설','공동주택', '노인복지시설', '녹지', '비상급수시설','지역아동센터','청소년공부방','어린이놀이시설'], axis=1, inplace=True)
print(data.info())


fig, ax = plt.subplots(len(data.columns.to_list()),1)
fig.set_size_inches(20,100)

for i, c in enumerate(data.columns.to_list()):
    sns.boxplot(x = data[c], ax = ax[i])
plt.show()
```

```python
def bar_and_box(data, column):
    fig, ax = plt.subplots(1, 2)
    fig.set_size_inches(20,8)
    sns.barplot(x = data['위험도2'], y = data[column], ax=ax[0])
    sns.boxplot(x = '위험도2', y = column, data = data, ax=ax[1])
    plt.savefig("C:/Users/gimm7/OneDrive/바탕 화면/cctv_분석.png")
    plt.show()
    return 

# bar_and_box(data, "경찰서")
```

```python
bar_and_box(data, 'cctv')
```
